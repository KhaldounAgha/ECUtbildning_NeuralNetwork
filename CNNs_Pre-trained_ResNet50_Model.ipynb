## Convolutional neural network CNNs

### The pre-trained ResNet50 model

```python

# Import necessary libraries and modules
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions  # Importing ResNet50 model and related functions for preprocessing and decoding predictions
from tensorflow.keras.preprocessing import image  # Importing image preprocessing module from TensorFlow
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions  # Importing preprocess_input and decode_predictions functions specifically from TensorFlow version of ResNet50
import numpy as np  # Importing numpy library for array manipulation
import matplotlib.pyplot as plt  # Importing matplotlib for plotting
```

```python

# Define an array containing image filenames
img_name_0 = np.array(['zebra.jpg', 'linjal.jpg', 'tennisboll.jpg', 'Green_Apple.jpg', 'Malkom_X.jpg', 'marilyn_monroe.jpg'])

# Create a subplot with 3 rows and 2 columns
fig, axes = plt.subplots(3, 2)

# Iterate through each subplot and corresponding image filename
for i, ax in enumerate(axes.flat):
    img_path = img_name_0[i]  # Get the image path
    image = plt.imread(img_path)  # Read the image using matplotlib
    ax.imshow(image)  # Display the image in the current subplot
    ax.set_title(img_path)  # Set the title of the subplot to the image filename

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python

# Load the pre-trained ResNet50 model with weights pre-trained on ImageNet dataset
model = ResNet50(weights='imagenet')

# Define an array containing image filenames
img_name_1 = np.array(['zebra.jpg', 'linjal.jpg', 'tennisboll.jpg', 'Green_Apple.jpg', 'Malkom_X.jpg', 'marilyn_monroe.jpg'])

# Iterate through each image filename in the array
for img_path in img_name_1:
    # Load and preprocess the image
    img = load_img(img_path, target_size=(224, 224))  # Load image and resize it to the required input size for ResNet50
    x = img_to_array(img)     # Convert image to numpy array
    x = np.expand_dims(x, axis=0)  # Expand dimensions to match the batch size expected by ResNet50
    x = preprocess_input(x)  # Preprocess the input image data

    # Make predictions using the pre-trained ResNet50 model
    preds = model.predict(x)  # Predict the class probabilities for the input image
    # Decode the predictions and print the top 3 predicted classes with their probabilities
    print(f'Imagefile to be predicted: {img_path} - Prediction: {decode_predictions(preds, top=3)[0]}')
    print('---')  # Print a separator between predictions for 
```

```python

# Load the pre-trained ResNet50 model with weights pre-trained on ImageNet dataset
model = ResNet50(weights='imagenet')

# Define an array containing image filenames
img_name_2 = np.array(['zebra.jpg', 'linjal.jpg', 'tennisboll.jpg', 'Green_Apple.jpg', 'Malkom_X.jpg', 'marilyn_monroe.jpg'])

# Iterate through each image filename in the array
for img_path in img_name_2:
    # Load and preprocess the image
    img = load_img(img_path, target_size=(224, 224))  # Load image and resize it to the required input size for ResNet50
    x = img_to_array(img)  # Convert image to numpy array
    print(x.shape)  # Print the shape of the image array
    x = x.reshape(1, 224, 224, 3)  # Reshape the image array to match the input shape expected by ResNet50
    x = preprocess_input(x)  # Preprocess the input image data

    # Make predictions using the pre-trained ResNet50 model
    preds = model.predict(x)  # Predict the class probabilities for the input image

    # Decode the predictions and print the top 3 predicted classes with their probabilities
    print(f'Imagefile to be predicted: {img_path} - Prediction: {decode_predictions(preds, top=3)[0]}')
    print('---')  # Print a separator between predictions for different images
```
